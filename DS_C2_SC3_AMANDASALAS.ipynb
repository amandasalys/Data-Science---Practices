2. Conjunta de manera correcta la información, de tal forma que queden todos los registros en un solo DataFrame que contenga todas las columnas necesarias.

```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

test1_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Test.csv')
test2_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Test2.csv')
train1_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Train.csv')
train2_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Train2.csv')

train_df = pd.merge(train1_df, train2_df, on='Employee_ID', how='left')
test_df = pd.merge(test1_df, test2_df, on='Employee_ID', how='left')


result = pd.concat([train_df, test_df], ignore_index=True)

print("")
result
```

3. Para todos los registros que no tienen un valor para un sistema operativo, ya sea de su computadora o de su dispositivo móvil, se asigna el valor “no registrado“.

```python
# Para cada valor Nan en las columnas de Computer_OS y Mobile_OS, se les asigna/sustituye el valor de "No registrado"

result = pd.concat([train_df, test_df], ignore_index=True)

result[' Computer_OS'] = result[' Computer_OS'].fillna('No registrado') #pd está leyendo NaN como float
result[' Mobile_OS'] = result[' Mobile_OS'].fillna('No registrado')


print("")
result
```

4. Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:

El porcentaje de personas que usan Windows, Mac, Linux o "No registrado" como su sistema operativo en su computadora. Se esperan 4 renglones en esta respuesta.

```python
import numpy as np

result = pd.concat([train_df, test_df], ignore_index=True)

result[' Computer_OS'] = result[' Computer_OS'].fillna('No registrado') #pd está leyendo NaN como float
result[' Mobile_OS'] = result[' Mobile_OS'].fillna('No registrado')

# Consultamos cuántos usuarios son los que manejan cada sistema operativo
c_so = result.groupby(' Computer_OS').size()

# Con la cantidad de usuario podemos calcular el porcentaje con las siguientes operaciones
# Aquí usamos .iloc para acceder a la posición de del df p_so
p_mac = (c_so.iloc[0] / c_so.sum()) * 100
p_win = (c_so.iloc[1] / c_so.sum()) * 100
p_lin = (c_so.iloc[2] / c_so.sum()) * 100
p_nr = (c_so.iloc[3] / c_so.sum()) * 100

p_so = {
    'Sistema Operativo': ['MacOS', 'Windows', 'Linux', 'No registrado'],
    'Porcentaje de usuarios que lo usan (%)': [p_mac, p_win, p_lin, p_nr]
}

p_so_df = pd.DataFrame(p_so)

print("")
p_so_df
```

5. Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:

El porcentaje de personas que usan iOS, Android o "No registrado" como su sistema operativo en su dispositivo móvil. Se esperan 3 renglones en esta respuesta.

```python
result = pd.concat([train_df, test_df], ignore_index=True)

result[' Computer_OS'] = result[' Computer_OS'].fillna('No registrado') #pd está leyendo NaN como float
result[' Mobile_OS'] = result[' Mobile_OS'].fillna('No registrado')

# Consultamos cuántos usuarios son los que manejan cada sistema operativo
c_so = result.groupby(' Mobile_OS').size()

# Con la cantidad de usuario podemos calcular el porcentaje con las siguientes operaciones
# Aquí usamos .iloc para acceder a la posición de del df p_so
p_ios = (c_so.iloc[0] / c_so.sum()) * 100
p_and = (c_so.iloc[1] / c_so.sum()) * 100
p_nr = (c_so.iloc[2] / c_so.sum()) * 100

p_som = {
    'Sistema Operativo': ['iOS', 'Android', 'No registrado'],
    'Porcentaje de usuarios que lo usan (%)': [p_ios, p_and, p_nr]
}

p_som_df = pd.DataFrame(p_som)

print("")
p_som_df
```

6. Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:

La edad promedio de la gente que utiliza cada sistema operativo de escritorio.

```python
result = pd.concat([train_df, test_df], ignore_index=True)

result[' Computer_OS'] = result[' Computer_OS'].fillna('No registrado') #pd está leyendo NaN como float
result[' Mobile_OS'] = result[' Mobile_OS'].fillna('No registrado')

# Buscamos cuando en COmputer_OS NO SEA No Registrado
usr_so = result[result[' Computer_OS'] != 'No registrado']

prom_edad = usr_so.groupby(' Computer_OS')['Age'].mean()

print("Edad promedio para cada Computer OS (excluyendo 'No registrado'):")
display(prom_edad)
```

7. Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:

El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio.

```python
prom_ne = usr_so.groupby(' Computer_OS')['Education_Level'].mean()

print("Nivel educativo promedio para cada Computer OS (excluyendo 'No registrado'):")
display(prom_ne)
```

8. Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:

El número total de tickets generados por la gente (Computer_tickets), con cada sistema operativo de escritorio.

```python
t_tickets = usr_so.groupby(' Computer_OS')[' Computer_tickets'].sum()

print("Total de Tickets generados para cada Computer OS (excluyendo 'No registrado'):")
display(t_tickets)
```
