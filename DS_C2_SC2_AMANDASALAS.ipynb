3. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
import pandas as pd

dataframe = pd.read_csv("/content/country_vaccinations.csv", parse_dates=["date"])

# Hacemos un df sobre los tipos de empresa
Can_Vac = dataframe.assign(empresa=dataframe["vaccines"].str.split(', ')).explode("vaccines")

# Agrupamos y sumamos para cada compañía su cantidad diaria de vacunas
Can_VaccxEmpresa = Can_Vac.groupby("vaccines")["daily_vaccinations"].sum()


Can_VaccxEmpresa
```

4. Obtener la cantidad de vacunas aplicadas en todo el mundo.

5. Calcular el promedio de vacunas aplicadas por país.

```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv("/content/country_vaccinations.csv", parse_dates=["date"])

# Hacemos un df sobre los  países
Can_Vac = dataframe.assign(country=dataframe["country"].str.split(', ')).explode("country")

# Agrupamos y obtenemos el promedio para cada país con la cantidad diaria de vacunas
Can_VaccxPais = Can_Vac.groupby("country")["daily_vaccinations"].mean()


Can_VaccxPais
```

6. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv("/content/country_vaccinations.csv", parse_dates=["date"])

# Filtramos para la fecha dada creando otro df
Vac_Fecha=dataframe[dataframe["date"]=="2021-01-29"]

# Calculamos la cantidad de vacunas total para la fecha dada
Vac_Total = Vac_Fecha["daily_vaccinations"].sum()

print("Total de vacunas aplicada para el 2021-01-29: ", Vac_Total)
```

11. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv("/content/country_vaccinations.csv", parse_dates=["date"])

pfizer = dataframe[dataframe["vaccines"].str.contains("Pfizer")]

pfizer
```

12. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv("/content/country_vaccinations.csv", parse_dates=["date"])

conDiferencias = dataframe.assign(diferencias = dataframe["daily_vaccinations"] - dataframe["daily_vaccinations_raw"])

conCantidad = dataframe.assign(canVac = dataframe["vaccines"].str.split(',').str.len())

antes20 = dataframe.loc[dataframe["date"]<"2020-12-20"]

pfizer = dataframe[dataframe["vaccines"].str.contains("Pfizer")]

with pd.ExcelWriter("resultadosReto.xlsx") as writer:
  conDiferencias.to_excel(writer, sheet_name="conDiferencias")
  conCantidad.to_excel(writer, sheet_name="conCantidad")
  antes20.to_excel(writer, sheet_name="antes20")
  pfizer.to_excel(writer, sheet_name="pfizer")
```
